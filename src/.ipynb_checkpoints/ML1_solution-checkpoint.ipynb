## Part 1. Introduction

### 1.1 Five examples of ML applications in real life

1. **Email spam filtering** — a classifier is trained on millions of labeled emails to automatically separate spam from legitimate mail. Benefit: saves time and protects users from phishing and malware.

2. **Medical diagnosis** — models analyze patient symptoms, lab results, and scan images (e.g., MRI) to predict diseases such as cancer. Benefit: faster and more consistent diagnosis, especially in areas with doctor shortages.

3. **Recommendation systems (Netflix, Spotify)** — collaborative filtering models learn from user behavior to suggest movies, songs, or products. Benefit: increases user satisfaction and platform engagement.

4. **Credit scoring in banking** — models predict the probability of loan default based on financial history and socioeconomic features. Benefit: reduces human bias, speeds up approval, and lowers financial risk for banks.

5. **Self-driving cars** — a combination of object detection (cameras, lidar) and decision-making models that navigate the vehicle in real time. Benefit: reduces human error, which is the leading cause of road accidents.

### 1.3 Difference between multiclass and multilabel

- **Multiclass classification**: each sample belongs to exactly **one** class out of several possible classes. The classes are mutually exclusive. Example: classifying a photo as 'cat', 'dog', or 'bird'.

- **Multilabel classification**: each sample can belong to **several classes at the same time**. Example: a news article can be tagged simultaneously with 'politics', 'economy', and 'Europe'.

The key difference: in multiclass, exactly one label is assigned per sample; in multilabel, zero or more labels can be assigned.

### 1.4 Is house price prediction classification or regression? Can regression be reduced to classification?

Predicting house prices is a **regression** problem, because the target variable (price) is continuous — it can take any numeric value.

However, it **is possible to reduce regression to classification** by discretizing the target variable into bins. For example: 'cheap' (< \$150k), 'medium' (\$150k–\$300k), 'expensive' (> \$300k). The model then predicts which category a house falls into.

The trade-off: we lose prediction precision (we no longer know the exact price), but the problem becomes simpler and sometimes more robust when the exact value is not critical.

## Part 3. Statistical Data Analysis

### 3.3 Target analysis

```python
# 3.3.1 Histogram of price (raw)
plt.figure(figsize=(10, 5))
plt.hist(df_small['price'], bins=100, color='steelblue', edgecolor='white')
plt.title('Distribution of Price (raw)')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.show()

print('The distribution is extremely right-skewed.')
print('Almost all data is compressed near zero because of extreme outliers (max = $4,490,000).')
print('The histogram is nearly unreadable in this form — outliers must be removed first.')
```

```python
# 3.3.2 Boxplot
plt.figure(figsize=(10, 4))
plt.boxplot(df_small['price'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='steelblue', color='steelblue'),
            medianprops=dict(color='white', linewidth=2))
plt.title('Boxplot of Price')
plt.xlabel('Price ($)')
plt.show()

print(f'Min:             ${df_small["price"].min():,.0f}')
print(f'Max:             ${df_small["price"].max():,.0f}')
print(f'Median:          ${df_small["price"].median():,.0f}')
print(f'1st percentile:  ${df_small["price"].quantile(0.01):,.0f}')
print(f'99th percentile: ${df_small["price"].quantile(0.99):,.0f}')
print()
print('The boxplot confirms massive outliers on both ends.')
print('The IQR box is invisible at this scale — the outliers dominate the plot completely.')
```

```python
# 3.3.4 Histogram after removing outliers
plt.figure(figsize=(10, 5))
plt.hist(df_small['price'], bins=80, color='seagreen', edgecolor='white')
plt.title('Distribution of Price (after removing outliers)')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.show()

print('After removing the top and bottom 1%, the distribution is much more readable.')
print('It is still right-skewed — most listings are at lower prices, with a long tail to the right.')
print('The bulk of NYC rental listings fall between $1,475 and $13,000 per month.')
```

## Part 4. Creating Features

```python
# 4.1 Add squared features
df_small['bathrooms_squared']      = df_small['bathrooms'] ** 2
df_small['bedrooms_squared']       = df_small['bedrooms'] ** 2
df_small['interest_level_squared'] = df_small['interest_level'] ** 2

corr_ext = df_small.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_ext, annot=True, fmt='.2f', cmap='coolwarm',
            square=True, linewidths=0.5)
plt.title('Correlation Matrix with Squared Features')
plt.tight_layout()
plt.show()

print('Correlation with price (sorted):')
print(corr_ext['price'].sort_values(ascending=False))
print()
print('The squared features show similar correlation with price as the original features.')
print('In this dataset the non-linear transformation does not dramatically improve correlation.')
print('This suggests the relationship between room counts and price is approximately linear.')
```

```python
# 4.2 For model training: use only 'bathrooms' and 'bedrooms'
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

X = df_small[['bathrooms', 'bedrooms']]
y = df_small['price']

# 4.4 Split: 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=21
)

print(f'X_train: {X_train.shape}  |  y_train: {y_train.shape}')
print(f'X_test:  {X_test.shape}   |  y_test:  {y_test.shape}')
```

```python
# 4.5-4.6 Apply PolynomialFeatures with degree=10
poly = PolynomialFeatures(degree=10)

X_train_poly = poly.fit_transform(X_train)  # fit on train only
X_test_poly  = poly.transform(X_test)       # transform test without fitting

print(f'Original features: 2')
print(f'After PolynomialFeatures(degree=10): {X_train_poly.shape[1]} features')
print(f'X_train_poly shape: {X_train_poly.shape}')
print(f'X_test_poly shape:  {X_test_poly.shape}')
```

---
## Part 5. Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')

def rmse(y_true, y_pred):
    """Root Mean Squared Error"""
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
# 5.1 Create empty result tables
result_MAE  = pd.DataFrame(columns=['model', 'train', 'test'])
result_RMSE = pd.DataFrame(columns=['model', 'train', 'test'])

print('Empty result tables created:')
print(result_MAE)
```

```python
# 5.2 Linear Regression

# 5.2.1 Initialize
lr = LinearRegression()

# 5.2.2 Fit and predict
lr.fit(X_train_poly, y_train)
lr_train_pred = lr.predict(X_train_poly)
lr_test_pred  = lr.predict(X_test_poly)

# Save predictions as new columns
X_train_res = X_train.copy()
X_test_res  = X_test.copy()
X_train_res['lr_prediction'] = lr_train_pred
X_test_res['lr_prediction']  = lr_test_pred

# 5.2.3 MAE
lr_mae_train = mean_absolute_error(y_train, lr_train_pred)
lr_mae_test  = mean_absolute_error(y_test,  lr_test_pred)

# 5.2.4 RMSE
lr_rmse_train = rmse(y_train, lr_train_pred)
lr_rmse_test  = rmse(y_test,  lr_test_pred)

print(f'Linear Regression — MAE  train: {lr_mae_train:.2f}, test: {lr_mae_test:.2f}')
print(f'Linear Regression — RMSE train: {lr_rmse_train:.2f}, test: {lr_rmse_test:.2f}')

# 5.2.5 Insert into tables
result_MAE  = pd.concat([result_MAE,  pd.DataFrame([{'model': 'linear_regression', 'train': lr_mae_train,  'test': lr_mae_test}])],  ignore_index=True)
result_RMSE = pd.concat([result_RMSE, pd.DataFrame([{'model': 'linear_regression', 'train': lr_rmse_train, 'test': lr_rmse_test}])], ignore_index=True)
```

```python
# 5.3 Decision Tree

# 5.3.1 Initialize with random_state=21
dt = DecisionTreeRegressor(random_state=21)

# 5.3.2 Fit and predict
dt.fit(X_train_poly, y_train)
dt_train_pred = dt.predict(X_train_poly)
dt_test_pred  = dt.predict(X_test_poly)

X_train_res['dt_prediction'] = dt_train_pred
X_test_res['dt_prediction']  = dt_test_pred

# 5.3.3 MAE
dt_mae_train = mean_absolute_error(y_train, dt_train_pred)
dt_mae_test  = mean_absolute_error(y_test,  dt_test_pred)

# 5.3.4 RMSE
dt_rmse_train = rmse(y_train, dt_train_pred)
dt_rmse_test  = rmse(y_test,  dt_test_pred)

print(f'Decision Tree — MAE  train: {dt_mae_train:.2f}, test: {dt_mae_test:.2f}')
print(f'Decision Tree — RMSE train: {dt_rmse_train:.2f}, test: {dt_rmse_test:.2f}')

# 5.3.5 Insert into tables
result_MAE  = pd.concat([result_MAE,  pd.DataFrame([{'model': 'decision_tree', 'train': dt_mae_train,  'test': dt_mae_test}])],  ignore_index=True)
result_RMSE = pd.concat([result_RMSE, pd.DataFrame([{'model': 'decision_tree', 'train': dt_rmse_train, 'test': dt_rmse_test}])], ignore_index=True)
```

```python
# 5.4 Naive Models

# 5.4.1 Calculate mean and median on TRAIN data, apply to both train and test
train_mean   = y_train.mean()
train_median = y_train.median()

nm_train_pred   = np.full(len(y_train), train_mean)
nm_test_pred    = np.full(len(y_test),  train_mean)
nmed_train_pred = np.full(len(y_train), train_median)
nmed_test_pred  = np.full(len(y_test),  train_median)

X_train_res['naive_mean_pred']   = nm_train_pred
X_test_res['naive_mean_pred']    = nm_test_pred
X_train_res['naive_median_pred'] = nmed_train_pred
X_test_res['naive_median_pred']  = nmed_test_pred

# 5.4.2-5.4.3 MAE and RMSE
nm_mae_train    = mean_absolute_error(y_train, nm_train_pred)
nm_mae_test     = mean_absolute_error(y_test,  nm_test_pred)
nm_rmse_train   = rmse(y_train, nm_train_pred)
nm_rmse_test    = rmse(y_test,  nm_test_pred)

nmed_mae_train  = mean_absolute_error(y_train, nmed_train_pred)
nmed_mae_test   = mean_absolute_error(y_test,  nmed_test_pred)
nmed_rmse_train = rmse(y_train, nmed_train_pred)
nmed_rmse_test  = rmse(y_test,  nmed_test_pred)

print(f'Naive Mean   — MAE  train: {nm_mae_train:.2f},   test: {nm_mae_test:.2f}')
print(f'Naive Mean   — RMSE train: {nm_rmse_train:.2f},  test: {nm_rmse_test:.2f}')
print(f'Naive Median — MAE  train: {nmed_mae_train:.2f},  test: {nmed_mae_test:.2f}')
print(f'Naive Median — RMSE train: {nmed_rmse_train:.2f}, test: {nmed_rmse_test:.2f}')

# 5.4.4 Insert into tables
result_MAE = pd.concat([result_MAE, pd.DataFrame([
    {'model': 'naive_mean',   'train': nm_mae_train,   'test': nm_mae_test},
    {'model': 'naive_median', 'train': nmed_mae_train, 'test': nmed_mae_test},
])], ignore_index=True)

result_RMSE = pd.concat([result_RMSE, pd.DataFrame([
    {'model': 'naive_mean',   'train': nm_rmse_train,   'test': nm_rmse_test},
    {'model': 'naive_median', 'train': nmed_rmse_train, 'test': nmed_rmse_test},
])], ignore_index=True)
```

```python
# 5.5.1 Print final result tables
print('=== result_MAE ===')
print(result_MAE.to_string(index=False))
print()
print('=== result_RMSE ===')
print(result_RMSE.to_string(index=False))
```

```python
# Visualize comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, result_df, metric in zip(axes, [result_MAE, result_RMSE], ['MAE', 'RMSE']):
    x = np.arange(len(result_df))
    width = 0.35
    ax.bar(x - width/2, result_df['train'].astype(float), width, label='Train', color='steelblue')
    ax.bar(x + width/2, result_df['test'].astype(float),  width, label='Test',  color='salmon')
    ax.set_xticks(x)
    ax.set_xticklabels(result_df['model'], rotation=20, ha='right')
    ax.set_title(f'{metric} by Model')
    ax.set_ylabel(metric)
    ax.legend()

plt.tight_layout()
plt.show()
```

### 5.5.2 Which model is the best?

| Model | MAE train | MAE test | RMSE train | RMSE test |
|-------|-----------|----------|------------|-----------|
| linear_regression | 756.73 | 759.83 | 1079.07 | 1247.30 |
| decision_tree | 0.00 | 753.99 | 0.00 | 1073.88 |
| naive_mean | 1140.45 | 1136.62 | 1598.46 | 1594.39 |
| naive_median | 1087.46 | 1081.22 | 1645.46 | 1639.34 |

**Analysis:**

- **Decision Tree** achieves MAE = 0 and RMSE = 0 on training data — this is classic **overfitting**. The model has memorized every training sample perfectly. The gap between train (0) and test (753) is the overfitting gap.

- **Linear Regression** is stable: train and test metrics are close (756 vs 759 MAE), which means the model generalizes well without overfitting.

- **Naive models** (mean, median) serve as the baseline. Both real models beat them by a large margin — proving they actually learn useful patterns.

- **Best model by test MAE**: Decision Tree (753.99) slightly edges out Linear Regression (759.83). However, Linear Regression is more reliable and interpretable in practice, with no overfitting. In a real project, we would prefer Linear Regression or add constraints (max_depth) to the Decision Tree.
